```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("result.txt") as file:
    raw = file.read()

string_list = raw.splitlines()

x = 10 * np.array([int(char) for char in string_list[0].split()])
y = np.array([float(item) for item in string_list[1].split()])

plt.plot(x, y)
plt.xlabel("MB")
plt.ylabel("sec")
plt.show()
print("Mean speed: %.3f MB/sec" % np.mean((x/y)))
```
