```python
import h5py
import numpy as np
from matplotlib import pyplot
import pandas
plt.style.use('ggplot')
plt.rcParams.update({'font.size': 14})
%matplotlib inline
```

```python
def plot_boxplot(df, title):
    #f, ax  = plt.figure(figsize = (15, 5))
    ax = df.plot.box(figsize = (15, 5), return_type = 'axes')
    ax.set_title(title)
    ax.set_ylabel('Mb/s')
    plt.show()
```

## AES

CBC mode

```python
with h5py.File('CBC_mode/128_enc.hdf5', 'r') as f:
    speed = f['speed']
    CBC128_enc = speed[:]
    
with h5py.File('CBC_mode/128_dec.hdf5', 'r') as f:
    speed = f['speed']
    CBC128_dec = speed[:]
    
with h5py.File('CBC_mode/192_enc.hdf5', 'r') as f:
    speed = f['speed']
    CBC192_enc= speed[:]
    
with h5py.File('CBC_mode/192_dec.hdf5', 'r') as f:
    speed = f['speed']
    CBC192_dec = speed[:]
    
with h5py.File('CBC_mode/256_enc.hdf5', 'r') as f:
    speed = f['speed']
    CBC256_enc = speed[:]
    
with h5py.File('CBC_mode/256_dec.hdf5', 'r') as f:
    speed = f['speed']
    CBC256_dec = speed[:]
# with ni
with h5py.File('CBC_mode/128_enc_ni.hdf5', 'r') as f:
    speed = f['speed']
    CBC128_enc_ni = speed[:]
    
with h5py.File('CBC_mode/128_dec_ni.hdf5', 'r') as f:
    speed = f['speed']
    CBC128_dec_ni = speed[:]
    
with h5py.File('CBC_mode/192_enc_ni.hdf5', 'r') as f:
    speed = f['speed']
    CBC192_enc_ni = speed[:]
    
with h5py.File('CBC_mode/192_dec_ni.hdf5', 'r') as f:
    speed = f['speed']
    CBC192_dec_ni = speed[:]
    
with h5py.File('CBC_mode/256_enc_ni.hdf5', 'r') as f:
    speed = f['speed']
    CBC256_enc_ni = speed[:]
    
with h5py.File('CBC_mode/256_dec_ni.hdf5', 'r') as f:
    speed = f['speed']
    CBC256_dec_ni = speed[:]
```

```python
CBC_mode_lbls = ['128(enc)', '128_(dec)', '192_(enc)',
          '192_(dec)','256_(enc)', '256_(dec)']

CBC_ni_dt = np.vstack((CBC128_enc_ni, CBC128_dec_ni, CBC192_enc_ni,
                       CBC192_dec_ni, CBC256_enc_ni, CBC256_dec_ni))
CBC_ni_df = pandas.DataFrame(CBC_ni_dt.transpose(), columns = CBC_mode_lbls)

CBC_dt = np.vstack((CBC128_enc, CBC128_dec, CBC192_enc,
                    CBC192_dec, CBC256_enc, CBC256_dec))
CBC_df = pandas.DataFrame(CBC_dt.transpose(), columns = CBC_mode_lbls)
```

```python
plot_boxplot(CBC_ni_df, title = 'Cipher Block Chaining(ni)')
```

```python
plot_boxplot(CBC_df, title = 'Cipher Block Chaining')
```
